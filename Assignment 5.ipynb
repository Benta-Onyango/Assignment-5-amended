Show the age distribution in the data using a histogram

```python
import plotly.graph_objs as go
```

```python
#data defining
trace = go.Histogram(x=titanic['Age'],opacity=0.75, name = "Age", histnorm='percent')
data = [trace]
# data layout
layout = go.Layout(title="Age Distribution",xaxis=dict(title='Age'),
                   yaxis=dict( title='Count'),)
fig = go.Figure(data=data, layout=layout)
fig.show()
```

Show the age distribution based on gender using a histogram.

```python
#data defining
trace = go.Histogram(x=titanic['Age'],opacity=0.75, name = "Age", histnorm='percent')
data = [trace]
# data layout
layout = go.Layout(title="Age Distribution based on Gender",xaxis=dict(title='Age'),
                   yaxis=dict( title='Sex'),)
fig = go.Figure(data=data, layout=layout)
fig.show()
```

Graph the value counts of the number of passengers who survived and did not survive
based on gender using a stacked bar graph.

```python
titanic['Survived'].value_counts()
```

```python
x0 = titanic['Survived'].value_counts()
x1 = titanic['Survived'].value_counts() + 1

fig = go.Figure()
fig.add_trace(go.Histogram(x=x0))
fig.add_trace(go.Histogram(x=x1))

# The two histograms are drawn on top of another
fig.update_layout(barmode='stack')
fig.show()
```
